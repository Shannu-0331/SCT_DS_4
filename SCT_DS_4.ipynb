```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
df = pd.read_csv(r'C://Users//tamid.SHANNU//OneDrive//Documents//Downloads//US_Accidents_March23.csv')
df.head(10)
```

```python
df.info()
print(f"Dataset contains {df.shape[0]} rows and {df.shape[1]} columns")
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.drop_duplicates(inplace=True)
df.dropna(subset=['Weather_Condition', 'Start_Time'], inplace=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Start_Time'] = pd.to_datetime(df['Start_Time'], errors='coerce')
df.dropna(subset=['Start_Time'], inplace=True)
df['Hour'] = df['Start_Time'].dt.hour
df['Day'] = df['Start_Time'].dt.day_name()
df['Month'] = df['Start_Time'].dt.month_name()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 4))
sns.countplot(x='Hour', data=df, palette='coolwarm')
plt.title('Accidents by Hour of the Day')
plt.xlabel('Hour')
plt.ylabel('Accident Count')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.countplot(x='Day', data=df, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], palette='magma')
plt.title('Accidents by Day of the Week')
plt.xlabel('Day')
plt.ylabel('Accident Count')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 4))
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']

sns.countplot(x='Month', data=df, order=month_order, palette='cubehelix')
plt.title('Accidents by Month')
plt.xlabel('Month')
plt.ylabel('Accident Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 4))
df['Weather_Condition'].value_counts().head(10).plot(kind='bar', color='blue')
plt.title('Top 10 Weather Conditions During Accidents')
plt.xlabel('Weather Condition')
plt.ylabel('Accident Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df.sample(1000), x='Start_Lng', y='Start_Lat', alpha=0.3)
plt.title('Sample Accident Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 4))
df['State'].value_counts().head(10).plot(kind='bar', color='green')
plt.title('Top 10 States with Highest Accidents')
plt.xlabel('State')
plt.ylabel('Accident Count')
plt.grid(True)
plt.show()
```
